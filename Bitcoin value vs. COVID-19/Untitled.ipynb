```python
import pandas as pd
import numpy as np
```

### Historical data about BTC price in USD$

```python
btc_dataset = pd.read_csv("BTC-USD.csv")
btc_dataset.head()
```

```python
btc_filtered = btc_dataset[(btc_dataset.Date >= '2019-31-12') & (btc_dataset.Date <= '2020-16-03')][["Date", "Open"]]
btc_filtered
```

```python
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

dates = btc_filtered.Date
x_btc = [dt.datetime.strptime(d,'%Y-%m-%d').date() for d in dates]
y_btc = btc_filtered.Open

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval = 10))
plt.plot(x_btc,y_btc)

plt.gcf().autofmt_xdate()
```

### Historical data about COVID-19 cases worldwide

```python
covid_dataset = pd.read_excel('COVID-19.xls')
covid_dataset.head()
```

```python
grouped_covid = covid_dataset.groupby("DateRep").sum()
grouped_covid.head(10)
```

```python
y = []
i = 0
j = 0

for i in range(len(grouped_covid.NewConfCases)):
    j = 0
    val = 0
    while(j <= i):
        val+=grouped_covid.NewConfCases[j]
        j+=1
    y.append(val)
```

```python
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

dates = grouped_covid.index
x = [dt.datetime.strptime(str(d.date()),'%Y-%m-%d').date() for d in dates]

plt.figure(figsize=(40,10))
fig = plt.figure()
fig, (axs1, axs2) = plt.subplots(1, 2, figsize = (15, 8))

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval = 10))

axs1.plot_date(x, y, 'b-')
axs1.title.set_text('Cases of COVID-19 worldwide')


axs2.plot_date(x_btc, y_btc, 'r-')
axs2.title.set_text('Bitcoin price in USD')


fig.autofmt_xdate()

plt.show()
```
