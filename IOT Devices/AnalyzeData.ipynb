### ** 2. How outside temperature was related to inside temperature? any relation between the two?**

```python
out_temp = dataset.loc[dataset["out/in"] == "Out"]
in_temp = dataset.loc[dataset["out/in"] == "In"]
```

```python
in_temp.temp
```

```python
import seaborn as sns
from scipy.stats import pearsonr

corr, _ = pearsonr(out_temp.temp[:len(in_temp.temp)], in_temp.temp)
print(corr)
```

### ** 3. which was the hottest/coolest month ?** 

```python
dataset["month"] = [date[3:5] for date in dataset["noted_date"]]
```

```python
temperatures_per_month = dataset.groupby(["month"]).mean()
```

```python
print("The hottest temperature is: " + str(temperatures_per_month.temp.max()))
```

```python
print("The coolest temperature is: " + str(temperatures_per_month.temp.min()))
```
